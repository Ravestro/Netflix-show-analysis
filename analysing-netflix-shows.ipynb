```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

reading datafile

```python
df = pd.read_csv("/kaggle/input/netflix-files/netflix_titles.csv")
```

```python
netflix = df.copy()
```

```python
netflix.dropna(inplace = True)
netflix.shape
```

```python
netflix["date_added"] = pd.to_datetime(netflix["date_added"])
netflix["day_added"] = netflix["date_added"].dt.day
netflix["month_added"] = netflix["date_added"].dt.month
netflix["year_added"] = netflix["date_added"].dt.year
netflix["day_added"].astype(int)
netflix["year_added"].astype(int)
netflix["month_added"].astype(int)

```

**Data Visualization**

```python
sns.countplot(data = netflix, x = 'type')
plt.title('Types')
```

```python
sns.countplot(data = netflix,x = "rating")
plt.xticks(rotation = 90)
plt.title('Ratings')
```

```python
sns.countplot(x='rating',hue='type',data = netflix)
plt.xticks(rotation = 90)
plt.title("Relation between type and rating")
```

Pie chart for type 

```python
plt.figure(figsize=(9, 9))
netflix['type'].value_counts().plot.pie(explode = [0,0.1])
```

Piechart for rating

```python
netflix["rating"].value_counts().plot.pie(autopct = '%1.1f%%')
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(
                          background_color='white',
                          width=1920,
                          height=1080
                         ).generate(" ".join(netflix['country']))
plt.imshow(wordcloud)
plt.axis('off')

```

```python
castcloud = WordCloud(background_color = "White",height = 1300,width = 1800).generate(" ".join(netflix.cast))
plt.imshow(castcloud)
plt.axis('off')
```

```python
directorcloud = WordCloud(background_color = "White",height = 1300,width = 1800).generate(" ".join(netflix.director))
plt.imshow(directorcloud)
plt.axis('off')
```

```python
categorycloud = WordCloud(background_color = "White",height = 1300,width = 1800).generate(" ".join(netflix.listed_in))
plt.imshow(categorycloud)
plt.axis('off')
```
